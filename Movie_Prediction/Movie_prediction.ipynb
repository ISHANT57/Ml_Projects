# 🎭 STEP 5 — Extract Top 3 Actors

```python
import ast

def convert2(obj):

    # agar string hai tab literal_eval karo
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)

    L = []
    for i in obj[:3]:
        if isinstance(i, dict):
            L.append(i['name'])
        else:
            L.append(i)

    return L
```

```python
print(type(movies['cast'][0]))
print(movies['cast'][0])
```

```python
movies['cast'] = movies['cast'].apply(convert2)
movies.head()
```

# 🎬 STEP 6 — Extract Director Only

```python
def convert3(obj):
    L=[]
    for i in eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(convert3)
movies.head()
```

# 🔹7 Text Cleaning

* Lowercase

* Remove spaces from names (Sam Worthington → samworthington)

* Stopwords remove (the, is, are, etc.)

* Stemming apply

```python
def remove_space(L):
    L1=[]
    for i in L:
        word=i.replace(" ","")
        L1.append(word)
    return L1
```

```python
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
```

```python
movies.head()
```

# 🏗 STEP 8 — Create Tags Column

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
data = [['Action','Hero'], ['Romance','Love']]

# Join
data = [" ".join(i) for i in data]

# Lowercase
data = [i.lower() for i in data]

print(data)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
# list ko string me convert kar diya
# List ke words ko ek single sentence me convert karna.
# List → String (sentence)

new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

# 🧬 5️⃣ Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
# It reduces words to their root form

# Removes suffix like ing, ed, ly, es, etc
```

```python
# text → full sentence string
```

```python
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python

new_df['tags'] = new_df['tags'].apply(stem)
new_df.head()
```

# VECTORIZATION

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
print(vectors)
print(cv.get_feature_names_out())
```

# 📐 6️⃣ Similarity Calculation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
print(similarity)
```

```python
print(new_df)
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
